```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files

print("ğŸ“ Upload a binary or simple object image:")
uploaded = files.upload()

for fn in uploaded.keys():
    image_path = fn
    break

# Load image (color for display, gray for processing)
img_color = cv2.imread(image_path)
gray = cv2.cvtColor(img_color, cv2.COLOR_BGR2GRAY)
gray = cv2.resize(gray, (512, 512))

# 1ï¸âƒ£ Thresholding â†’ binary image
_, binary = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

# 2ï¸âƒ£ Connected Component Labeling
num_labels, labels, stats, centroids = cv2.connectedComponentsWithStats(binary, connectivity=8)

# 3ï¸âƒ£ Assign random colors to each label
output = np.zeros((labels.shape[0], labels.shape[1], 3), dtype=np.uint8)
colors = [np.random.randint(0, 255, 3).tolist() for i in range(num_labels)]
for i in range(1, num_labels):  # skip background (label 0)
    output[labels == i] = colors[i]

# 4ï¸âƒ£ Compute shape descriptors
print(f"ğŸ§© Detected {num_labels-1} objects\n")

for i in range(1, num_labels):
    x, y, w, h, area = stats[i]
    cx, cy = centroids[i]
    roi = binary[y:y+h, x:x+w]
    perimeter = cv2.arcLength(cv2.findContours(roi, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)[0][0], True)
    aspect_ratio = float(w) / h
    circularity = 4 * np.pi * area / (perimeter**2 + 1e-6)

    print(f"Object {i}:")
    print(f"  â†’ Area           : {area}")
    print(f"  â†’ Centroid       : ({cx:.2f}, {cy:.2f})")
    print(f"  â†’ Bounding Box   : {w}x{h}")
    print(f"  â†’ Perimeter      : {perimeter:.2f}")
    print(f"  â†’ Aspect Ratio   : {aspect_ratio:.2f}")
    print(f"  â†’ Circularity    : {circularity:.3f}")
    print("-" * 40)

    # draw centroid marker
    cv2.circle(output, (int(cx), int(cy)), 5, (255, 255, 255), -1)

# 5ï¸âƒ£ Display Results
plt.figure(figsize=(14,6))
plt.subplot(1,2,1)
plt.imshow(binary, cmap='gray')
plt.title('Binary Image')
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(cv2.cvtColor(output, cv2.COLOR_BGR2RGB))
plt.title(f'Labeled Objects: {num_labels-1}')
plt.axis('off')
plt.tight_layout()
plt.show()
```
