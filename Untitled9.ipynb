```python
# ============================================================
# UNIT 5: Photo Album with Face Detection & Recognition (OpenCV Only)
# ============================================================

import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files
import os

# ------------------------------------------------------------
# 1Ô∏è‚É£ Upload known face(s)
# ------------------------------------------------------------
print("üì∏ Upload 1‚Äì3 clear face images (used as 'known' people):")
known_uploads = files.upload()

known_faces = []
known_names = []

face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades +
                                     "haarcascade_frontalface_default.xml")

for fn in known_uploads.keys():
    img = cv2.imread(fn)
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(gray, 1.2, 5)
    if len(faces) > 0:
        x, y, w, h = faces[0]
        face_crop = gray[y:y+h, x:x+w]
        known_faces.append(face_crop)
        name = os.path.splitext(fn)[0]
        known_names.append(name)
        print(f"‚úÖ Added known face: {name}")
    else:
        print(f"‚ö†Ô∏è No face detected in {fn}")

if len(known_faces) == 0:
    raise ValueError("‚ùå No valid known faces uploaded!")

# ------------------------------------------------------------
# 2Ô∏è‚É£ Upload album photos
# ------------------------------------------------------------
print("\nüñºÔ∏è Upload album photos for recognition:")
album_uploads = files.upload()

# ------------------------------------------------------------
# 3Ô∏è‚É£ Recognize faces in each album photo
# ------------------------------------------------------------
def mse(img1, img2):
    img1 = cv2.resize(img1, (100, 100))
    img2 = cv2.resize(img2, (100, 100))
    return np.mean((img1.astype("float") - img2.astype("float")) ** 2)

for fn in album_uploads.keys():
    img = cv2.imread(fn)
    rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    faces = face_cascade.detectMultiScale(gray, 1.2, 5)
    for (x, y, w, h) in faces:
        face_crop = gray[y:y+h, x:x+w]

        # Compare with known faces using MSE (simplest metric)
        scores = [mse(face_crop, kf) for kf in known_faces]
        best_idx = int(np.argmin(scores))
        name = known_names[best_idx] if scores[best_idx] < 2000 else "Unknown"

        cv2.rectangle(rgb, (x, y), (x+w, y+h), (0, 255, 0), 2)
        cv2.putText(rgb, name, (x, y-10), cv2.FONT_HERSHEY_SIMPLEX,
                    0.6, (255, 0, 0), 2)

    plt.figure(figsize=(6,6))
    plt.imshow(rgb)
    plt.title(f"Detected faces in: {fn}")
    plt.axis('off')
    plt.show()
```
