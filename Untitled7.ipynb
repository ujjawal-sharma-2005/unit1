```python
# ============================================================
# UNIT 3: Line and Circle Detection using Hough & RANSAC
# ============================================================

import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files
from sklearn.linear_model import RANSACRegressor
from sklearn.datasets import make_regression

print("üìÅ Upload an image with clear edges or shapes:")
uploaded = files.upload()
for fn in uploaded.keys():
    image_path = fn
    break

# ------------------------------------------------------------
# 1Ô∏è‚É£ Load & Preprocess
# ------------------------------------------------------------
img_color = cv2.imread(image_path)
img_color = cv2.resize(img_color, (512, 512))
gray = cv2.cvtColor(img_color, cv2.COLOR_BGR2GRAY)
edges = cv2.Canny(gray, 80, 180)

# ------------------------------------------------------------
# 2Ô∏è‚É£ Line Detection using Hough Transform
# ------------------------------------------------------------
hough_img = img_color.copy()
lines = cv2.HoughLines(edges, 1, np.pi / 180, 120)
if lines is not None:
    for rho, theta in lines[:, 0]:
        a, b = np.cos(theta), np.sin(theta)
        x0, y0 = a * rho, b * rho
        x1, y1 = int(x0 + 1000 * (-b)), int(y0 + 1000 * (a))
        x2, y2 = int(x0 - 1000 * (-b)), int(y0 - 1000 * (a))
        cv2.line(hough_img, (x1, y1), (x2, y2), (0, 255, 0), 2)

# ------------------------------------------------------------
# 3Ô∏è‚É£ Circle Detection using Circular Hough Transform
# ------------------------------------------------------------
circles_img = img_color.copy()
circles = cv2.HoughCircles(gray, cv2.HOUGH_GRADIENT, dp=1.2,
                            minDist=40, param1=100, param2=30,
                            minRadius=20, maxRadius=150)
if circles is not None:
    circles = np.uint16(np.around(circles))
    for (x, y, r) in circles[0, :]:
        cv2.circle(circles_img, (x, y), r, (255, 0, 0), 3)
        cv2.circle(circles_img, (x, y), 2, (255, 255, 255), 3)

# ------------------------------------------------------------
# 4Ô∏è‚É£ Robust Line Fitting using RANSAC
# ------------------------------------------------------------
# Extract edge points (x, y)
y_idxs, x_idxs = np.nonzero(edges)
X = x_idxs.reshape(-1, 1)
y = y_idxs

ransac = RANSACRegressor(min_samples=2, residual_threshold=3)
ransac.fit(X, y)
line_y = ransac.predict(X)

ransac_img = img_color.copy()
for xi, yi in zip(x_idxs, line_y.astype(int)):
    if 0 <= yi < 512:
        ransac_img[int(yi), xi] = (0, 0, 255)

# ------------------------------------------------------------
# 5Ô∏è‚É£ Display Results
# ------------------------------------------------------------
plt.figure(figsize=(14, 10))

plt.subplot(2, 2, 1)
plt.imshow(cv2.cvtColor(img_color, cv2.COLOR_BGR2RGB))
plt.title("Original")
plt.axis('off')

plt.subplot(2, 2, 2)
plt.imshow(edges, cmap='gray')
plt.title("Canny Edges")
plt.axis('off')

plt.subplot(2, 2, 3)
plt.imshow(cv2.cvtColor(hough_img, cv2.COLOR_BGR2RGB))
plt.title("Hough Transform Lines")
plt.axis('off')

plt.subplot(2, 2, 4)
plt.imshow(cv2.cvtColor(circles_img, cv2.COLOR_BGR2RGB))
plt.title("Circular Hough Transform")
plt.axis('off')

plt.tight_layout()
plt.show()

plt.figure(figsize=(6,6))
plt.imshow(cv2.cvtColor(ransac_img, cv2.COLOR_BGR2RGB))
plt.title("RANSAC Line Fitting")
plt.axis('off')
plt.show()
```
